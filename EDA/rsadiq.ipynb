```python
covid_df = covid_df.dropna(subset=['total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'population']).drop_duplicates(subset=['location'], keep='last')
from IPython.display import HTML, display
HTML(covid_df.to_html())



```

```python
covid_df.columns = ['Location', 'Collection Date', 'Total Vaccinations', 'People Vaccinated', 'People Fully Vaccinated', 'Population']

```

Percentage fully Vaccinated

```python
covid_df['Percentage Fully Vaccinated'] = covid_df['People Fully Vaccinated'] / covid_df['Population'] * 100
covid_df
```

```python
countries_data = covid_df[covid_df['Location'].isin(countries)]
continents_data = covid_df[~covid_df['Location'].isin([*countries, 'World'])]

len(countries_data), len(continents_data)
```

```python
def plot_bars(data, x_label, y_label, k=10, plot_type='total_vaccinations'):
    if k:
        data_slice = data[:k]
    else:
        data_slice = data[:20]
        
    # set tooltip list
    tooltip_list = ['Collection Date', 'Population', 'Location']
    if plot_type == 'total_vaccinations':
        tooltip_list.append('Total Vaccinations')
    elif plot_type == 'fully_vaccinated':
        tooltip_list.append('People Fully Vaccinated')
        tooltip_list.append('Percentage Fully Vaccinated')
    elif plot_type == 'people_vaccinated':
        tooltip_list.append('People Vaccinated')

    bars = alt.Chart(data_slice).mark_bar().encode(
        alt.Y(y_label, sort=None),
        x=x_label,
        tooltip = tooltip_list,
    )
#     .interactive().configure_axis(grid=False)

#     text = bars.mark_text(
#         align='left',
#         baseline='middle',
#         dx=5  # Nudges text to right so it doesn't appear on top of the bar
#     ).encode(text=x_label)

    return (bars)
```

## Total Vaccinations Administered

### Top X countries

```python
total_vaccinations_by_countries = countries_data.sort_values(by=['Total Vaccinations'], ascending=False)
total_vaccinations_by_countries
```

```python
plot_bars(total_vaccinations_by_countries, 'Total Vaccinations:Q', 'Location:O', k=10, plot_type = 'fully_vaccinated')
```

### Total Vaccinations administered per continent

```python
total_vaccinations_by_continents = continents_data.sort_values(by=['Total Vaccinations'], ascending=False)
total_vaccinations_by_continents
```

```python
plot_bars(total_vaccinations_by_continents,  'Total Vaccinations:Q', 'Location:O')
```
